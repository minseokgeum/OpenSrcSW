```python
import csv
f = open("seoul.csv", 'r', encoding="cp949")
data = csv.reader(f, delimiter=',')
header = next(data)
max_temp = -999
max_date = ""
min_temp = 999
min_date = ""
max_gap = 0
gap_temp_h = 0
gap_temp_l = 0

gap_date = ""

for row in data:
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    gap = row[-1] - row[-2]
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
    if min_temp > row[-2]:
        min_date = row[0]
        min_temp = row[-2]
    if max_gap < gap:
        gap_date = row[0]
        gap_temp_h = row[-1]
        gap_temp_l = row[-2]
        max_gap = gap
        
f.close()

print("2021년 서울의 최고 기온이 가장 높았던 날은 {:s}로, {:.1f}도 였습니다.".format(max_date, max_temp))
print("2021년 서울의 최저 기온이 가장 낮았던 날은 {:s}로, {:.1f}도 였습니다.".format(min_date, min_temp))
print("2021년 서울의 일교차가 가장 컸던 날은 {:s}로, {:.1f}도 였습니다. (최고 기온 : {:.1f}도, 최저 기온 : {:.1f}도)".format(gap_date, max_gap, gap_temp_h, gap_temp_l))
```
